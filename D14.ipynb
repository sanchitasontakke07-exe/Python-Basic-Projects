```python
import numpy as np
arr = np.random.randn(6, 6)
print("Array:",arr)
print("\nShape:",arr.shape)
print("\nSize:",arr.size)
print("\nData Type:",arr.dtype)
print("\nIndex of Maximum Value:",np.unravel_index(np.argmax(arr), arr.shape))
print("\nIndex of Minimum Value:",np.unravel_index(np.argmin(arr), arr.shape))
print("\nTop-left 3x3 Submatrix:",arr[:3, :3])
arr = np.abs(arr)
print("\nModified Matrix:",arr)
print("\nMean of Modified Matrix:",np.mean(arr))
```

```python
import numpy as np
marks=np.random.randint(30,101,(10,5))
print("Student Marks :",marks)
# Calculate total and average
total=np.sum(marks,axis=1)
average=np.mean(marks,axis=1)
print("Total Marks :",total)
print("Average Marks :",average)
# Find highest and lowest scorer
print("Highest Total Marks Student Index :",np.argmax(total))
print("Lowest Total Marks Student Index :",np.argmin(total))
print("Class Mean :",np.mean(marks))
print("Class Standard Deviation :",np.std(marks))
# Top 3 students
top3=np.argsort(total)[-3:]
print("Top 3 Students Marks :",marks[top3])
```
